```python
#Import Packages 

import pandas as pd

file_path ='nba_data.csv'

team_data = pd.read_csv(file_path)

print('CSV has successfully been read')

team_data.head()
```

```python
#Grouping data by 'Team' and calculating the sum of numerical columns

grouped_data = team_data.groupby('Team').sum(numeric_only=True) 
print (grouped_data)
```

```python
#Removing the 'Games Played' column from the team DataFrame.

team_summary = grouped_data.drop(columns=['Games Played']).groupby('Team').sum(numeric_only=True)
print(team_summary)
```

```python
#Adding a new column 'Games Played' with a value of 50 for all teams.

team_summary['Games Played'] = 50 
print (team_summary)
```

```python
#Calculating 'Pts per Game' (Points per Game) for each team and adding it as a new column to the team DataFrame.

team_summary['PTS per Game'] = round(team_summary['PTS'] / team_summary['Games Played'],1)
print (team_summary)
```

```python
#Graph 1: Showing which teams have the most three-pointers ('3P').

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.scatter(team_summary.index, team_summary['3P'], color='Blue', alpha=0.9)
plt.xlabel('Teams')
plt.ylabel('3Ps')
plt.title('Total Three-Pointers by Team')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Graph 2: Showing which teams have the most assists ('AST').

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.scatter(team_summary.index, team_summary['AST'], color='Green', alpha=0.9)
plt.xlabel('Teams')
plt.ylabel('AST')
plt.title('Total AST by Team')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Graph 3: Showing which teams have the most steals ('STL').

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.scatter(team_summary.index, team_summary['STL'], color='Red', alpha=0.9)
plt.xlabel('Teams')
plt.ylabel('STL')
plt.title('Total STL by Team')
plt.tight_layout()
plt.grid(True)
plt.show()
```
